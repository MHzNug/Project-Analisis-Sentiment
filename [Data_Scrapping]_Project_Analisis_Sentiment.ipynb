```python
from google_play_scraper import Sort, reviews
import pandas as pd
```

```python
# Mengambil Data Review Aplikasi
result, _ = reviews(
    'com.indomaret.klikindomaret',     # Ganti dengan ID aplikasi target
    lang='id',              # Bahasa Indonesia
    country='id',           # Negara Indonesia
    sort=Sort.MOST_RELEVANT,       # Bisa NEWEST, RATING, atau HELPFUL
    count=15000,             # Jumlah review yang ingin diambil
)

# Melihat hasil
for r in result[:5]:
    print(f"[{r['score']}★] {r['content']}")
```

```python
# Melihat DataFrame
df = pd.DataFrame(result)
df.head()
```

```python
df.to_csv('klikindomaret_reviews.csv', index=False)
```
